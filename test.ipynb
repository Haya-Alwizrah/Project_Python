```python
bord = [
    ["X", "X", "X"],
    [" ", " ", " "],
    ["O", "O", "O"]
]
```

```python
def print_bord(bord):
    for i in bord:
        print(i)

print_bord(bord)
```

```python
def pos(bord:list[list]):
    x_pos = []
    o_pos = []
    for i, r in enumerate(bord):
        for j, c in enumerate(r):
            if c == "X":
                x_pos.append((i,j))
            elif c == "O":
                o_pos.append((i,j))
    return x_pos, o_pos

x_pos, o_pos = pos(bord)
print(x_pos)
print(o_pos)
```

```python
def norm_win(bord:list[list])->bool:
    if "X" in bord[2] or "O" in bord[0]:
        return True
    else:
        return False
    
print(norm_win(bord))
```

```python
def inside(i, j):
    return 0 <= i < 3 and 0 <= j < 3
```

```python
player = "X"

def a_move(player:str, bord):
    x_pos, o_pos = pos(bord)
    if player == "X":
        m = 1
        p = x_pos
        e = "O"
    else:
        m = -1
        p = o_pos
        e = "X"

    a_move = []

    for i,j in p:
        # قدام
        new_i, new_j = i+m, j
        if inside(new_i, new_j) and bord[new_i][new_j] == " ":
            a_move.append(((i, j), (new_i, new_j)))

        # يهاجم
        for z in (1,-1):
            new_i, new_j = i+m, j+z
            if inside(new_i, new_j) and bord[new_i][new_j] == e:
                a_move.append(((i, j), (new_i, new_j)))

    return a_move

m = a_move(player, bord)
for i in m:
    print(i)
```
